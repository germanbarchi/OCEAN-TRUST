```python
import pandas as pd
```

```python
df_features=pd.read_csv('data/features/new_partitions-egemaps_all_audio.csv')
```

```python
for i in df_features.index:
    df_features.loc[i,'basename']=df_features.loc[i,'Name'].split('.')[0]
```

```python
quantile_df=df_features.groupby('basename').quantile(.5).reset_index()
quantile_df=pd.merge(quantile_df,df_features.loc[:,['Name','basename']],on='basename')
```

```python
median_df=df_features.groupby('basename').mean().reset_index()
median_df=pd.merge(median_df,df_features.loc[:,['Name','basename']],on='basename')
```

```python
df_features
```

```python
a=pd.merge(df_features,median_df,on='Name',suffixes=('','_(median)'))
df=pd.merge(a,quantile_df,on='Name',suffixes=('','_(quantile)'))
```

```python
features=df_features.columns[~df_features.columns.isin(['Name','Part','basename'])]
for i in features:
    df[i]=df[i]-df[i+'_(median)']
    df[i]=df[i]/df[i+'_(quantile)']
```
