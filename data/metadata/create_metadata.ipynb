```python
import pandas as pd
```

```python
dev=pd.read_csv('eth_gender_annotations_dev.csv')
test=pd.read_csv('eth_gender_annotations_test.csv')
metadata=pd.read_csv('metadata_df.csv',index_col=0)
```

```python
df=pd.concat([dev,test],ignore_index=True)
```

```python
import pandas as pd
import warnings
df_final=pd.DataFrame()

for i in df.index:
    split_data=df.loc[i,df.columns.values[0]].split(';')
    df_final.loc[i,'filename']=split_data[0].replace('.mp4','.wav')
    df_final.loc[i,'basename']=split_data[1]
    df_final.loc[i,'ethnicity']=split_data[2]
    df_final.loc[i,'gender']=split_data[3]
```

```python
df_final.to_csv('metadata_ethnicity_gender.csv')
```
