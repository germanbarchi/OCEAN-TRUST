```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Find correlation between loudness and equivalent level and personality traits

```python
features=pd.read_csv('../data/features/new_partitions-egemaps_all_audio.csv')
labels=pd.read_csv('../data/metadata/metadata_df.csv')
```

```python
df_all=pd.merge(features,labels,left_on='Name',right_on='filename')
cols=['loudness_sma3_amean','equivalentSoundLevel_dBp','basename', 'extraversion', 'neuroticism',
       'agreeableness', 'conscientiousness','openness','labels_mean']
```

```python
df_=df_all[cols]
sns.heatmap(df_.corr(), annot=True)
plt.show()
```

Min max normalization

```python
norm_features=pd.read_csv('../data/features/egemaps_all_audios_normalized.csv')
df_norm_all=pd.merge(norm_features,labels,left_on='Name',right_on='filename')
```

```python
df_norm=df_norm_all[cols]
```

```python
sns.heatmap(df_norm.corr(), annot=True)
plt.show()
```
